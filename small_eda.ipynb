```python
import pandas as pd
df = pd.read_parquet('data.parquet')
for col in df.columns:
    print(col)
    print('unique count is : ', df[col].nunique())
    print('unique sample  is : ', df[col].unique().tolist()[:5])
    if not df[col].dtype=='O':
        print('stats of columns is : ' ,)
        print('avg is  : ' ,df[col].mean())
        print('min is  : ' ,df[col].min())
        print('max is  : ' ,df[col].max())
        print('std is  : ' ,df[col].std())
    
    print('-------------')
    print('\n')
    
```
